# Import Packages

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Get data

```python
# Euler Solutions
E10 = pd.read_csv("Euler_10.csv", index_col=0)
E20 = pd.read_csv("Euler_20.csv", index_col=0)
E40 = pd.read_csv("Euler_40.csv", index_col=0)
E80 = pd.read_csv("Euler_80.csv", index_col=0)

# RK Solutions
RK10 = pd.read_csv("RK_10.csv", index_col=0)
RK20 = pd.read_csv("RK_20.csv", index_col=0)
RK40 = pd.read_csv("RK_40.csv", index_col=0)
RK80 = pd.read_csv("RK_80.csv", index_col=0)

# Exact Solution
t = np.linspace(0,1,100)
y = .5*t**2 + t + 2
```

```python
ax = E10.plot()
E20.plot(ax=ax)
E40.plot(ax=ax)
E80.plot(ax=ax)

ax = RK10.plot()
RK20.plot(ax=ax)
RK40.plot(ax=ax)
RK80.plot(ax=ax)
plt.plot(t,y)
```
